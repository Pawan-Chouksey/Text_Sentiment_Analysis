```python
import numpy as np
import pandas as pd
import re
```

```python
data=pd.read_csv('IMDB.csv')
```

```python
def preprocess(text):
    if not isinstance(text, str):
        return ''
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)
    text = re.sub(r'\@\w+|\#','', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = text.lower()
    return text

data['cleaned_text'] = data['text'].apply(preprocess)

data.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords') #not required if you used jupyter notebook in native compiler having nltk library
nltk.download('wordnet')   #not required if you used jupyter notebook in native compiler having nltk library
lemmatizer = WordNetLemmatizer()
def remove(text):
    tokens = text.split()
     #compile words same meaning with different form of words h.
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stopwords.words('english')]
    return ' '.join(tokens) #return a string

data['cleaned_text'] = data['cleaned_text'].apply(remove)

data.head()
```

```python
data.to_csv('Clean_Data.csv', index=False)
```
